```python
# Importing required libraries
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt

# Import custom classes
from imdb import Imdb
from reuters import Reuters
from boston_housing import BostonHousing
```

## 1. Binary Classification: IMDB Movie Reviews

```python
# IMDB Binary Classification
print("\n--- IMDB Movie Review Sentiment Classification ---")

# Initialize IMDB class
imdb_classifier = Imdb()

# Prepare data
x_train_imdb, x_test_imdb, y_train_imdb, y_test_imdb = imdb_classifier.prepare_data()
print(f"IMDB Training data shape: {x_train_imdb.shape}")
print(f"IMDB Training labels shape: {y_train_imdb.shape}")

# Build model
imdb_model = imdb_classifier.build_model()
imdb_model.summary()
```

```python
# Train the IMDB model
imdb_history = imdb_classifier.train(epochs=10)

# Plot loss and accuracy
imdb_classifier.plot_loss()
imdb_classifier.plot_accuracy()

# Evaluate the model
imdb_test_loss, imdb_test_accuracy = imdb_classifier.evaluate()
```

## 2. Multiclass Classification: Reuters Newswire

```python
# Reuters Multiclass Classification
print("\n--- Reuters Newswire Classification ---")

# Initialize Reuters class
reuters_classifier = Reuters()

# Prepare data
x_train_reuters, x_test_reuters, y_train_reuters, y_test_reuters = reuters_classifier.prepare_data()
print(f"Reuters Training data shape: {x_train_reuters.shape}")
print(f"Reuters Training labels shape: {y_train_reuters.shape}")

# Build model
reuters_model = reuters_classifier.build_model()
reuters_model.summary()
```

```python
# Train the Reuters model
reuters_history = reuters_classifier.train(epochs=10)

# Plot loss and accuracy
reuters_classifier.plot_loss()
reuters_classifier.plot_accuracy()

# Evaluate the model
reuters_test_loss, reuters_test_accuracy = reuters_classifier.evaluate()
```

## 3. Regression: Boston Housing Prices

```python
# Boston Housing Regression
print("\n--- Boston Housing Price Prediction ---")

# Initialize Boston Housing class
boston_regressor = BostonHousing()

# Prepare data
x_train_boston, x_test_boston, y_train_boston, y_test_boston = boston_regressor.prepare_data()
print(f"Boston Housing Training data shape: {x_train_boston.shape}")
print(f"Boston Housing Training labels shape: {y_train_boston.shape}")

# Build model
boston_model = boston_regressor.build_model()
boston_model.summary()
```

```python
# Train the Boston Housing model
boston_history = boston_regressor.train(epochs=100)

# Plot loss
boston_regressor.plot_loss()

# Evaluate the model
boston_test_loss, boston_test_mae = boston_regressor.evaluate()
```
